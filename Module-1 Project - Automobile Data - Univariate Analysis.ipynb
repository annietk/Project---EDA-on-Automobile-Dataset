```python
corr = autocl.corr()
plt.figure(figsize=(20,9))
a = sns.heatmap(corr, annot=True, fmt='.2f')
```

```python
From the heat map, we can see that:
    a) the price is positively correlated to wheel-base,length,width,curb-weight,engine-size,horse-power and number of cylinders
    b)symboling is negatively correlated to most of the attributes except for normalized-losses
    c)All the physical attributes of a vehicle should be positively correlated and the map really says it so by the way the
    following attributes like wheel-base, length,width,curb-weight,engine-size are distributed.
    d)highway-mpg and city-mpg are highly correlated.
    e)Also, the engine-size and horse-power are also correlated in a positive way.
```

```python
autocl['volume'] = autocl['length']*autocl['height']*autocl['width']
```

```python
autocl.info()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pylab 

fig = plt.figure(figsize=(12,10))

ax = fig.add_subplot(111, projection='3d')


x = autocl['num-doors']
y = autocl['num-cyl']
z = autocl['volume']
c = autocl['price']


cb1=ax.scatter(x, y, z, c=c, cmap=plt.cm.get_cmap('RdBu'),s=100)
plt.colorbar(cb1, ax=ax)
ax.set_xlabel('No. of Doors',fontsize=20)
ax.set_ylabel('No. of Cylinders',fontsize=20)
ax.set_zlabel('Volume',fontsize=20)


plt.show();
```

```python
pd.pivot_table(autocl,index=['body-style','num-of-doors','drive-wheels'], values = 'normalized-losses').plot(kind='bar');
```

```python
Here , we can see that the normalized losses(relative average loss payment per insured vehicle year) is 
high for two-doors vehicles than four-doors vehicle but its distributed across different body style and drive-wheels type.
Also we can see that normalized-losses is comparetively more for rwd(rear-wheel drive) which makes sense as rwd is found often
in sports cars, performance sedans, big trucks, race cars, and law enforcement pursuit vehicles. 
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pylab 

fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(111, projection='3d')


x = autocl['engine-size']
y = autocl['horsepower']
z = autocl['compression-ratio']
c = autocl['num-cyl']


cb1=ax.scatter(x, y, z, c=c, cmap=plt.cm.get_cmap('seismic'),s=100)
plt.colorbar(cb1, ax=ax)
ax.set_xlabel('Engine Size',fontsize=20)
ax.set_ylabel('Horsepower',fontsize=20)
ax.set_zlabel('compression-ratio',fontsize=20)
plt.show();
```
